## Conjuntos

```python
# Un conjunto (set)es una colección que no tiene orden ni repetición, colección no ordenada de valores únicos
# A|B (Unión): elementos en A o en B
# A-B (Diferencia): elementos en A no presentes en B = A intersección Bcomplemento
# Diferencia simetrica: elementos en cualquiera de los conujntos A o B, más no en los dos
# A <= B (Subconjunto): si todos los elementos en A se encuentran en B
```

```python
vocales = {'a', 'e', 'i', 'o', 'u'}
p = {'c', 'a', 'r', 'o'}
```

```python
type(vocales)
```

```python
# Intersección
p&vocales
```

```python
# Union
p|vocales
```

## Diccionario

1. Un diccionario es una colección no ordenada de pares: llave - valor
2. Es MUTABLE
3. Las keys (llaves) DEBEN SER únicas, los valores NO

```python
# Ejemplo
menu = {
    "sopa": 89, 
    "salmón": 259,
    "atún": 319,
    "pechuga pollo" : 189
}
# La longitud de un diccionario es equivalente al número de pares dentro del mismo
```

```python
type(menu)
```

```python
len(menu)
```

```python
menu["salmón"]
```

```python
menu
```

```python
# Si quiero agregar un elemento, debo hacerlo llave <-> valor
menu["espárragos"] = 200
```

```python
menu
```

```python
# Si quiero quitar un valor, al igual que las listas,usamos el método pop()
menu.pop("espárragos")
```

```python
menu
```

```python
'leche' in menu
```

```python
'salmón' in menu 
```

```python
200 in menu
```
